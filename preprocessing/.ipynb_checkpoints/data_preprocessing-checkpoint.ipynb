```python
import pandas as pd
import os
import urllib
import pdb
import cv2

LABEL_PATH = '../data/label/label.xls'
IMAGE_ROOT = '../data/image'
FRONT_PATH = '../data/image_front'
BACK_PATH = '../data/image_back'
```

```python
def download(label_path, output_root):
    ''' Download images based on image link
    '''
    xls = pd.read_excel(label_path)
    for i in range(len(xls['link'])):
        link = xls['link'][i]
        name = xls['No'][i]
        img_name = os.path.join(output_root, str(name)) + '.jpg'
        if not(isinstance(link, str)):
            continue
#         if link.split(':')[0] == 'http':
        try:
            urllib.request.urlretrieve(link, img_name)
        except Exception as e:
            print(link, e)
    

def imgCrop(img_root):
    ''' Croping and Resize images to 1024*512*3
    '''
    garbage = []
    for idx, img_name in enumerate(os.listdir(img_root)):
        path = os.path.join(img_root, img_name)
        img = cv2.imread(path)
        
        try:
            img.size
        except Exception as e:
            print(type(img))
            print(img_name)
            garbage.append(img_name)
            #print(garbage)
            continue
            
        if not img.size == 1024*512*3:
            img = cv2.resize(img, (1300,710))
            img = img[10:10+650, :, :]
            img = cv2.resize(img, (1024,512))
            cv2.imwrite(path, img)
            print(idx)
        
    for i in garbage:
        file = os.path.join(img_root, i)
        if os.path.isfile(file):
            os.remove(file)
            print('remove' + file)
           
            
def sizeCheck(img_root):
    size_list = []
    for img_name in os.listdir(img_root):
        path = os.path.join(img_root, img_name)
        img = cv2.imread(path)
        if isinstance(img, type(None)):
            print(path)
        else:
            size_list.append(img.shape)
    print(set(size_list))

    
def imgCut(img_root, front_path, back_path):
    for idx, img_name in enumerate(os.listdir(img_root)):
        path = os.path.join(img_root, img_name)
        tmp = cv2.imread(os.path.join(img_root, img_name), cv2.IMREAD_COLOR)
        img_front = tmp.copy()
        img_front = tmp[0:512, 0:512]
        img_back = tmp.copy
        img_back = tmp[0:512, 513:1024]
        
        cv2.imwrite(os.path.join(front_path,img_name), img_front)
        cv2.imwrite(os.path.join(back_path,img_name), img_back)
        
        if idx%100 == 0:
            print(idx)
#         cv2.imshow("tmp", tmp)
#         cv2.imshow("img-front", img_front)
#         cv2.imshow("img-back", img_back)
#         cv2.waitKey(0)
#         cv2.destroyAllWindows()
            

# download(LABEL_PATH, IMAGE_ROOT)
# imgCrop(IMAGE_ROOT)
# sizeCheck(IMAGE_ROOT)
# imgCut(IMAGE_ROOT, FRONT_PATH, BACK_PATH)
```
