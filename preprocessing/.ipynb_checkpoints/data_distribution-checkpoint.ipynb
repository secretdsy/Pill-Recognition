```python
csv = pd.read_csv(os.path.join(LABEL_PATH, 'shape_color_prediction.csv'))
df = pd.DataFrame(csv)
```

```python
df.head()
```

```python
shape_unique = len(df['shape'].unique().tolist())
color_unique = len(df['color_front'].unique().tolist())
```

```python
shape_list = df['shape'].tolist()
shape_count_list = []

for i in range(shape_unique):
    shape_count_list.append(shape_list.count(i))
print(shape_count_list)
```

```python
shape_list = df['shape_top1'].tolist()
shape_count_list = []

for i in range(shape_unique):
    shape_count_list.append(0)
    shape_count_list[i] = shape_list.count(i)
print(shape_count_list)
```

```python
color_list = df['color_front'].tolist()
color_count_list = []

for i in range(color_unique):
    color_count_list.append(color_list.count(i))
print(color_count_list)
```

```python
color_list = df['color_top1'].tolist()
color_count_list = []

for i in range(color_unique):
    color_count_list.append(0)
    color_count_list[i] = color_list.count(i)
print(color_count_list)
```

```python
label_list = [[0 for _ in range(color_unique)] for _ in range(shape_unique)]

for i in range(len(df)):
    for j in range(shape_unique):
        for k in range(color_unique):
            if df['shape'][i] == j and df['color_front'][i] == k:
                label_list[j][k] += 1
```
